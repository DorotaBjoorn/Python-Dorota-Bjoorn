```python
# GRUNDUPPGIFT

import matplotlib.pyplot as plt
import numpy as np
import math

with open ("../Labs/Data_labs/Lab02_datapoints.txt", "r") as data_file:
    # Reading in data where each line is a string in a list. [1:] removes the header line
    datapoints_list_initial = data_file.readlines()[1:]         

    # crating data_point_list: [[width, hight, label], [width, hight, label], ...]
    data_point_list = []                                         
    for line in datapoints_list_initial:
        (width, hight, label) = line.split(",")
        width = float(width)
        hight = float(hight)
        label = int(label)
        data_point_list.append([width, hight, label])
    
    # changing list to an array
    data_point_array = np.array(data_point_list)

    # filtering into 2 separate arrays to be able to plot as separate series and thus get legends
    pikachu = data_point_array[data_point_array[:,2] == 1] # whole original array[filtering condition]
    pichu = data_point_array[data_point_array[:,2] == 0]

    plt.scatter(pichu[:,0], pichu[:,1])
    plt.scatter(pikachu[:,0], pikachu[:,1])
    plt.legend(["Pichu", "Pikachu"])
    plt.title("Pichu and Pikachu scatter")
    plt.xlabel("pokemon width")
    plt.ylabel("pokemon hight")


with open("../Labs/Data_labs/Lab02_testpoints.txt", "r") as test_file:
    # Reading in data where each line is a string in a list. [1:] removes the header line
    testpoints_list_initial = test_file.readlines()[1:] 

    # creating a test point list [[width, hight], [width, hihgt], ...]
    test_point_list =[]
    for line in testpoints_list_initial:
        line = line.split("(")[1]                           # splits at ( and keeps everything that is after by indicating [1]
        line = line.split(")")[0]                           # splits at ) and keeps everything that is before by indicating [0]
        (width_test, hight_test) = line.split(",")          # splits at , and keeps what is before as width and after as hight
        width_test = float(width_test)
        hight_test = float(hight_test)
        test_point_list.append([width_test, hight_test])

    # creating a distance list for a test point [[distance, label], [distance, label], ...]. Once list is created it is sorted
    # and label corresponding to the top position (lowest distance) is appended to a label list.append
    # Same is repeted for all 4 test point resulting i a label list [label from test 1, label from test 2, ...]
    classification_list = []
    for width_test, hight_test in test_point_list:
        distance_list = []
        for width, hight, label in data_point_list:
            distance = math.sqrt((width_test-width)**2+(hight_test-hight)**2)
            distance_list.append([distance, label])
            distance_list.sort() 
        classification_list.append(distance_list[0][1])  # from sorted list picking label ([1]) from first row/lowest distance ([0])

    print(f"Result from current classification is {classification_list}, where expected result is [1, 1, 1, 0]")
```

```python
# EXERCISE 2

import matplotlib.pyplot as plt
import math

with open ("../Labs/Data_labs/Lab02_datapoints.txt", "r") as data_file:
    datapoints_list_initial = data_file.readlines()[1:]         

    # Gathering data points in a list [width, hight, label]
    data_point_list = []                                         
    for line in datapoints_list_initial:
        (width, hight, label) = line.split(",")
        width = float(width)
        hight = float(hight)
        label = int(label)
        data_point_list.append([width, hight, label])
    
    # Prompting user for input and checks so values are not letters and not below 0
    while True:
        user_width = input("Type in pokemon width")
        try:
            user_width = float(user_width)
            if user_width <= 0:
                print("Value should be granter than 0")
            else:
                break
        except:
            print("Numeric values only")

    while True:
        user_hight = input("Type in pokemon hight")
        try:
            user_hight = float(user_hight)
            if user_hight <= 0:
                print("Value should be granter than 0")
            else:
                break
        except:
            print("Numeric values only")
  
    distance_list = []
    classification = 0
    for width, hight, label in data_point_list:
        distance = math.sqrt((user_width-width)**2+(user_hight-hight)**2)
        distance_list.append([distance, label])
        distance_list.sort() 
    for i in range(0,10):
        classification = classification + distance_list[i][1]

    print(f"Your value for width {user_width} and hight {user_hight} result in classification {classification}")

    if classification > 5:
        print("Your Pokemon is most probably a Pikachu")
    elif classification < 5:
        print("Your Pokemon is most probably a Pichu")
    else:
        print("Your Pokemon can not be classified as Pichu or Pikachu")
```
