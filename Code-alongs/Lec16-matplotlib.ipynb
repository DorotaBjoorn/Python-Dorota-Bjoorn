# Matplotlib

```python
import matplotlib.pyplot as plt         #plt existerar i namespace
import numpy as np

# array from -10 to 10 with 100 numbers equally spaced
x = np.linspace(-10, 10, 20)           # skulle kunna kallas x_array då globalt namespace medan x i lambda är ett lokalt x så inte samma x

f = lambda x: np.cos(x)                 # tar bara 1 element från arrayen f=np.cos(x) skulle applicera funktionen på alla x i arrayen
# rad 6 är samma som
# def f(x):
    #return np.cos(x)
```

```python
x = np.linspace(-10, 10, 200)
plt.figure(figsize=(8,4), dpi = 50)
plt.plot(x, f(x))
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")

#h-line och v-line för att få axlar, plt.box(False) för att bli av med rutan runt
```

## OOP approach

```python
# instntiated fig, ax from figure and axes
fig, ax = plt.figure(figsize = (8,4)), plt.axes()
ax.plot(x, f(x))
ax.set(xlabel = "x", ylabel = "y", title = "$\cos{x}$", xlim = (-5,5), ylim = (-1.2, 1.2))
fig.savefig("../assets/Lec16_cos(x).png", dpi=200)




#help(plt)
```

#### Subplots

```python
# Matlab approach
plt.figure(figsize=(16,5))
plt.subplot(1,2,1) # (#rows, #columns, panel number)
plt.plot(x, f(x))
plt.plot(x, np.sin(x))
plt.title("cos(x), sin(x)")

plt.subplot(1,2,2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("Trigonometric functions", fontweight = "bold")
```

```python
# OOP approach
fig, ax = plt.subplots(1, 2, figsize =(16,5))

ax[0].plot(x, f(x))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))

ax[0].set(title = "cos(x) & sin(x)")
fig.suptitle("Big title", fontsize = 25)

# to access 2x2 -> ex 3rd ax: ax[1][0], 4e ax[1][1]
```

```python
plt.plot(x, f(x), 'o--') #x och y värden skickas in i var sin array (eller lista)
# kantig eftersom baseras på 20 punkter bara
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

data_normal = np.random.randn(1000)

ax[0].hist(data_normal, bins=20, alpha=0.7)
ax[0].set(title="Histogram of normally distributed data", xlabel="x", ylabel="y")


size = 1000 * np.random.randn(100)


scatter = ax[1].scatter(
    x=np.random.randn(100),
    y=np.random.randn(100),
    s=size,
    alpha=0.3,
    cmap="viridis",
    c=np.random.randn(100),
)

ax[1].set(title="Bubble chart", xlabel = "x", ylabel = "y");

fig.colorbar(scatter)
fig.set_facecolor("lightblue")

# alfa - hur genomskinligt
# set - kan sätta många parametrar på samma rad
# cmap - färgtema
# c - färg för viss datavärde
```
