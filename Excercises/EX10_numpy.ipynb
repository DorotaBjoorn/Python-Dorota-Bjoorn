```python
import numpy as np
import matplotlib.pyplot as plt

# generate arrey for number of rolls by comprehension (can not use append with arrays)
rolls = np.array([10**i for i in range (1,8)])
print(f"Arrey of rolls is {rolls}")

# generate empty array with 0, which are to be replaced with average values from the for loop below
average = np.zeros(7)

# loop through rolls array: generate average for each roll in rolls and replace 0 in average array with outcome at position i
for i, roll in enumerate(rolls): #returnerar index och roll # alt för for loopen: for i, roll in zip(range (0, len(rolls)), rolls):
    dice = np.random.randint(1,7, roll)
    average[i] = dice.mean()

print(f"Array of averages is {average}")

plt.xscale("log") # gör om x-axeln till logaritmisk skala
plt.plot(rolls, average, color = "green", marker = "o", linestyle = "dashed")
plt.title("Average of 6-sided die")
plt.xlabel("number of rolls")
plt.ylabel("average")
```

```python
# 2e Löst med Anders, men får inte till graferna
import numpy as np
import matplotlib.pyplot as plt

def dice_throws(number_of_dices):

    dice = np.arange(1,7)
    sum_matrix = np.zeros((6,)*number_of_dices)
    for i in range(number_of_dices):
        dim_list = [1]*i + [6] + [1]*(number_of_dices-1-i) # ger shape
        sum_matrix = sum_matrix + dice.reshape(dim_list)

    values, counts = np.unique(sum_matrix, return_counts = True)
    frequencies = counts/sum_matrix.size
    total_frequency = np.sum(frequencies)
    print(f"Total frequency should be 1 and it is = {total_frequency}")
    return(frequencies, values)

for i in range (1,5):
    frequencies, values = dice_throws (i)
    plt.bar(values, frequencies) # ger alla plottar i samma

frequencies_1, values_1 = dice_throws(1)
frequencies_2, values_2 = dice_throws(2)
frequencies_3, values_3 = dice_throws(3)
frequencies_4, values_4 = dice_throws(4)
fig, ((dice_1, dice_2), (dice_3, dice_4)) = plt.subplots(2,2) # ger 4 subplottar men med kontsing inzoomning
dice_1.bar(frequencies_1, values_1)
dice_2.bar(frequencies_2, values_2)
dice_3.bar(frequencies_3, values_3)
dice_4.bar(frequencies_4, values_4)
```
